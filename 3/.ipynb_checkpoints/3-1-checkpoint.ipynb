### - Mine the Association rules with a minimum Support of 1% and a minimum Confidence of 0%.
    c. How many rules appear in the data?
    d. How many rules are observed when the minimum confidence is 50%.
    e. Explain how the specified confidence impacts the number of rules.

```r
#Mine the Association rules with a minimum Support of 1% and a minimum Confidence of 0%
m_rules = apriori(data, parameter=list(support=0.01, confidence=0.0))
```

**c. How many rules appear on the data?** The number of rules can be seen above in writing ... [11524 rule(s)]

```r
#d. How many rules are observed when the minimum confidence is 50%.
m_rules = apriori(data, parameter=list(support=0.01, confidence=0.5))
```

1,165 rules with minimum confidence of 50%

**e. Explain how the specified confidence impacts the number of rules.** Since we specified the minimum confidence to be 50%, this will reduce the number of rules generated by only considering the transactions that have at least a pair of items that appear at least 50% of the time

### - Create a scatter plot comparing the parameters support and confidence on the axis, and lift with shading.

f. Identify the positioning of the interesting rules.

```r
m_rules <- apriori(data,parameter =list(supp=0.01, conf=0.0),control = list(verbose = FALSE))
plot(m_rules, jitter = 0)
```

The interesting rules are those rules that have high confidence and high lift
