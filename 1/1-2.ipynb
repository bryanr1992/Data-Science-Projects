We do not want the correlation of mpg and mpg or cylinders with cylinders. I did not know how to do this in R so I looked it up

```r
#___________________________________________________________
# This function was found online (by Stephen Turner)
#This is use to calculate the p-val and correlation value of the variables
#Remember that correlation does not imply causation

cor.prob <- function (X, dfr = nrow(X) - 2) {
  M <- cor(X, use="pairwise.complete.obs")
  above <- row(M) < col(M)
  r2 <- M[above]^2
  Fstat <- r2 * dfr/(1 - r2)
  M[above] <- 1 - pf(Fstat, 1, dfr)
  M[row(M) == col(M)] <- NA
  M 
}
```

```r
# Use this to dump the cor.prob output to a 4 column matrix
# with row/column indices, correlation, and p-value.
# Source: https://stackoverflow.com/questions/12116207/flatten-matrix-in-r-to-four-columns-indexes-and-upper-lower-triangles

flattenSquareMatrix <- function(m) {
  if( (class(m) != "matrix") | (nrow(m) != ncol(m))) stop("Must be a square matrix.") 
  if(!identical(rownames(m), colnames(m))) stop("Row and column names must be equal.")
  ut <- upper.tri(m)
  data.frame(i = rownames(m)[row(m)[ut]],
             j = rownames(m)[col(m)[ut]],
             cor=t(m)[ut],
             p=m[ut])
}
```

```r
cor.Matrix <- flattenSquareMatrix(cor.prob(df[,1:6]))
cor.Matrix
```

```r
# I also found a better way to show the relationships 
# from our scatter plots from stack overflow

panel.hist = function(x, ...) {
  usr = par("usr")
  on.exit(par(usr))
  par(usr = c(usr[1:2], 0, 1.5))
  h = hist(x, plot = F)
  breaks = h$breaks
  nB = length(breaks)
  y = h$counts
  y = y / max(y)
  rect(breaks[-nB], 0, breaks[-1], y, col = "cyan", ...)
}
panel.cor = function(x, y, digits = 2, prefix = "", cex.cor, ...) {
  usr = par("usr")
  on.exit(par(usr))
  par(usr = c(0, 1, 0, 1))
  r = cor(x, y)
  txt = format(c(r, 0.123456789), digits = digits)[1]
  txt = paste0(prefix, txt)
  if (missing(cex.cor)) {
    cex.cor = 0.8 / strwidth(txt)
  }
  text(0.5, 0.5, txt, cex = cex.cor * r)
}
```

```r
pairs(new_df[,1:6], lower.panel = panel.smooth, diag.panel = panel.hist, upper.panel = panel.cor)
```
