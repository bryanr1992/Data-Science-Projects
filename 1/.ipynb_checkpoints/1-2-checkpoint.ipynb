```r
df <-read.csv("Auto.csv")
View(df)
str(df)
```

```r
#Check for missing vals
colSums(is.na(df))
```

**a. Which of the predictors are quantitative, and which are qualitative?**

Qualitative variables are: name, Year, Origin

Quantitative: cylinders, mpg, displacement, horsepower, weight, acceleration

```r
#For convinience we are going to change the type of variables to numeric instead
#of int

df$cylinders <- as.numeric(df$cylinders)
df$horsepower <- as.numeric(df$horsepower )
df$year <- as.factor(df$year)
df$weight <- as.numeric(df$weight)
df$origin <- as.factor(df$origin)

str(df)
```

```r
#Check for missing data again
colSums(is.na(df))
```

```r
#Eliminate the missing data
new_df <- na.omit(df)
```

```r
colSums(is.na(new_df))
```
